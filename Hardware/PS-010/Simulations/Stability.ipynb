```python
import ltspice
import matplotlib.pyplot as plt
from matplotlib.ticker import EngFormatter
import numpy as np
import scipy
from nfft import nfft
```

```python
l = ltspice.Ltspice('PS-010.raw')
l.parse() # Data loading sequence. It may take few minutes for huge file.

f_lower = 1e3
f_upper = 1e6
```

```python
time = l.getTime()
V_out = l.getData('V(out)')
V_in = l.getData('V(in)')

plt.plot(time, V_in)
plt.plot(time, V_out)
plt.xlim(0,0.001)
plt.show()
```

```python
#resample data
xnew = np.linspace(np.min(time), np.max(time), len(time)*10, endpoint=False)
v_in_resampler = scipy.interpolate.interp1d(time, V_in)
v_out_resampler = scipy.interpolate.interp1d(time, V_out)
V_in_resamp=v_in_resampler(xnew)
V_out_resamp=v_out_resampler(xnew)
```

```python
%matplotlib inline
Fs=Fs=1/(xnew[1]-xnew[0])
fig, ax = plt.subplots()
inn = ax.magnitude_spectrum(V_in_resamp,Fs=Fs,scale='dB') 
outt = ax.magnitude_spectrum(V_out_resamp,Fs=Fs,scale='dB') 
formatter = EngFormatter(unit='Hz')
ax.xaxis.set_major_formatter(formatter)
ax.set_xscale('log')
plt.xlim(f_lower,f_upper)
plt.xlabel('Frequency (Hz)')
print()
```

```python
Fs=Fs=1/(xnew[1]-xnew[0])
fig, ax = plt.subplots()
inp = ax.phase_spectrum(V_in_resamp,Fs=Fs) 
outp = ax.phase_spectrum(V_out_resamp,Fs=Fs) 
formatter = EngFormatter(unit='Hz')
ax.xaxis.set_major_formatter(formatter)
ax.set_xscale('log')
plt.xlim(f_lower,f_upper)
plt.ylim(-360,np.pi)
plt.xlabel('Frequency (Hz)')
```

```python
%matplotlib inline
fig, ax = plt.subplots()
tf=outt[0]/inn[0]
index_lower=(np.abs(inn[1]-f_lower)).argmin()
index_upper=(np.abs(inn[1]-f_upper)).argmin()
mag_interp_inv=scipy.interpolate.interp1d(tf[index_lower:int(index_upper*.95)],inn[1][index_lower:int(index_upper*.95)])
ax.plot(inn[1],10*np.log10(tf))
#ax.plot(x,10*np.log10(asdf(x)))
formatter = EngFormatter(unit='Hz')
ax.xaxis.set_major_formatter(formatter)
ax.set_xscale('log')
plt.xlim(f_lower,f_upper)
plt.ylim(-16,10)
plt.xlabel('Frequency (Hz)')
plt.show()
print('-3dB: ',formatter(mag_interp_inv(1/np.sqrt(2))))
```
