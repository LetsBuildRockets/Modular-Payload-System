```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import EngFormatter
%matplotlib inline
csv = np.genfromtxt ('+12V Input Filter Simulation Results.csv', skip_header=1, delimiter=',')    
```

```python
fig, ax = plt.subplots()
ax.hist(csv[:,0]) 
formatter = EngFormatter(unit='Hz')
ax.xaxis.set_major_formatter(formatter)
plt.title('+12V Input Filter -3dB Cutoff Frequency Histogram')
plt.xlabel('-3dB Cutoff Frequency (Hz)')
plt.ylabel('Frequency')
plt.show()
print('Min cutoff Frequency:',formatter(np.min(csv[:,0])))
print('Avg cutoff Frequency:',formatter(np.average(csv[:,0])))
print('Max cutoff Frequency:',formatter(np.max(csv[:,0])))
```

```python
plt.hist(csv[:,1])
plt.title('+12V Input Filter Quality Factor Histogram')
plt.xlabel('Quality Factor')
plt.ylabel('Frequency')
plt.show()
print('Note: The methond cannot measure Q of < 1. Any Q of < 1 in the histogram will be repoted as 1.')
print('max Q:',np.max(csv[:,1]))
```

```python
plt.hist(csv[:,2])
plt.title('+12V Input Filter Impedance Histogram')
plt.xlabel('Z out (ohms)')
plt.ylabel('Frequency')
plt.show()
formatter = EngFormatter(unit='Ohms')
print('Min Z_out:',formatter(np.min(csv[:,2])))
print('Avg Z_out:',formatter(np.average(csv[:,2])))
print('Max Z_out:',formatter(np.max(csv[:,2])))
```
